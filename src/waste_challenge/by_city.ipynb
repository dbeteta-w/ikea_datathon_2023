```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
```

```python
DATA_PATH = '../1. Data/'
WASTE_PATH = DATA_PATH+'waste.csv'
```

```python
df = pd.read_csv('../1. Data/waste.csv', delimiter=';', decimal='.')
df['date'] = pd.to_datetime(df['date'])  # convert date column to datetime format
```

## By city analysis

```python
# total waste per city pie chart
df.groupby('location').sum()['waste'].plot(kind='pie', figsize=(5,5), autopct='%1.1f%%', title='Total waste per city')
```

```python
df_bcn = df[df['location']=='Barcelona']
df_md = df[df['location']=='Madrid']
```

```python
# total waste pie chart per food category
df_bcn.groupby('food_category').sum()['waste'].plot(kind='pie', figsize=(5,5), autopct='%1.1f%%', title='Total waste per food category in Barcelona')
```

```python
# madrid
df_md.groupby('food_category').sum()['waste'].plot(kind='pie', figsize=(5,5), autopct='%1.1f%%', title='Total waste per food category in Madrid')
```

```python
# waste per kitchen 
df_bcn.groupby('kitchen_type').sum()['waste'].plot(kind='pie', figsize=(5,5), autopct='%1.1f%%', title='Total waste per kitchen in Barcelona')
```

```python
df_md.groupby('kitchen_type').sum()['waste'].plot(kind='pie', figsize=(5,5), autopct='%1.1f%%', title='Total waste per kitchen in Madrid')
```
